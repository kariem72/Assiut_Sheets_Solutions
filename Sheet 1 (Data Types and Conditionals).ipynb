### 14- Calculator (N)

Given a mathematical expression in the form $A \ S \ B$ where $S$ is either $(+, -, \times, /)$. Print the result.

**Input**
One line containing A, S and B ($1 \le A, B \le 10^4$).

**Output**
Print the result of the mathematical expression.

```python
expression = input()

operators = ['+', '-', '*', '/']
op = ''

for char in expression:
    if char in operators:
        op = char
        break

A, B = map(int, expression.split(op))

if op == '+':
    print(A + B)
elif op == '-':
    print(A - B)
elif op == '*':
    print(A * B)
elif op == '/':
    # Integer division
    print(A // B)
```

### 20- Float or int (T)

Given a number N. Determine whether N is a float or integer. If it is a float, print the integer part and decimal part. If it is an integer, print only the integer part.

**Input**
Only one line containing a number N ($1 \le N \le 10^3$)

**Output**
Print the required format, e.g., `int 234` or `float 234 000`.

```python
N_str = input()

if '.' in N_str:
    integer_part, decimal_part = N_str.split('.')
    
    # Remove trailing zeros to see if it's truly a float
    decimal_part_trimmed = decimal_part.rstrip('0')
    
    if not decimal_part_trimmed:
        print(f"int {integer_part}")
    else:
        # Use the trimmed decimal part
        print(f"float {integer_part} {decimal_part_trimmed}")
else:
    print(f"int {N_str}")
```

### 21- Comparison (U)

Given a comparison $A \ S \ B$. Determine whether the comparison is Right or Wrong.

**Input**
One line containing A, S and B ($-100 \le A, B \le 100$), where S is ('<', '>', '=').

**Output**
Print "Right" if the comparison is true, "Wrong" otherwise.

```python
line = input().split()
A = int(line[0])
S = line[1]
B = int(line[2])

is_right = False

if S == '<' and A < B:
    is_right = True
elif S == '>' and A > B:
    is_right = True
elif S == '=' and A == B:
    is_right = True

if is_right:
    print("Right")
else:
    print("Wrong")
```

### 25- Hard Compare (Y)

Given 4 numbers A, B, C and D. If $A^B > C^D$ print "YES" otherwise, print "NO".

**Input**
One line containing 4 numbers A, B, C and D ($1 \le A, C \le 10^7$), ($1 \le B, D \le 10^{12}$)

**Output**
Print "YES" or "NO".

```python
import math

A, B, C, D = map(int, input().split())

# Direct calculation of A**B is impossible due to large exponents.
# We compare B * log(A) with D * log(C), because log(X^Y) = Y * log(X).

left_side = B * math.log(A)
right_side = D * math.log(C)

if left_side > right_side:
    print("YES")
else:
    print("NO")
```
